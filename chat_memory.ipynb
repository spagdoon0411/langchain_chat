```python
def chat_with_ollama(model=MODEL):
    print("ðŸ¦™ Ollama Chat | Type 'exit' to quit\n")

    chat_history = []  # Stores conversation context
    simple_history = []

    while True:
        try:
            user_input = input("You: ").strip()
            simple_history.append({"input": user_input})
            if user_input.lower() in {"exit", "quit"}:
                print("Goodbye! ðŸ‘‹")
                break

            chat_history.append({"role": "user", "content": user_input})
            response = ollama.chat(model=model, messages=chat_history)

            reply = response["message"]["content"]
            print(f"Ollama: {reply}\n")

            chat_history.append(
                {"role": "assistant", "content": reply}
            )  # Maintain context
            simple_history.append({"output": reply})

        except KeyboardInterrupt:
            print("\nGoodbye! ðŸ‘‹")
            break

    return simple_history, ollama_history
```

```python
simple_history, ollama_history = chat_with_ollama()
```
